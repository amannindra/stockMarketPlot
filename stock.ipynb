```python
import numpy as np
import pandas as pd
from pandas_datareader import data as pdr
import yfinance as yf
import matplotlib.pyplot as plt

# Ensure yfinance overrides pandas_datareader's download method
yf.pdr_override()

tickers = ['AAPL', 'GOOGL', 'MSFT', 'AMZN']

# Download stock data
data = yf.download(tickers, period='300d')

# Create a figure and a set of subplots
fig, axes = plt.subplots(nrows=len(tickers), ncols=1, figsize=(10, 15), sharex=True)

# Check if axes is an array, if not, make it one
if len(tickers) == 1:
    axes = [axes]

for ax, ticker in zip(axes, tickers):
    # Extract closing prices for the ticker
    df = data['Close'][ticker]
    
    # Calculate moving averages
    df_50_MA = df.rolling(window=50).mean()
    df_200_MA = df.rolling(window=200).mean()
    
    # Plot closing price
    ax.plot(df.index, df, label='Closing Price', color='black')
    
    # Plot moving averages
    ax.plot(df_50_MA.index, df_50_MA, label='50-Day MA', color='blue')
    ax.plot(df_200_MA.index, df_200_MA, label='200-Day MA', color='red')
    
    # Set the title and legend
    ax.set_title(ticker)
    ax.legend()

plt.tight_layout()
plt.savefig("simplefinance.png", dpi=200)
plt.show()
```
